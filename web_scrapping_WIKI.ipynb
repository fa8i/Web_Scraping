### WEB SCRAPING - WIKIPEDIA

Web scraping de tablas complejas en Wikipedia

### IMPORTACION LIBRERIAS

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd
```

### URL Y HEADERS

```python
url = 'https://es.wikipedia.org/wiki/Anne_Hathaway'
headers = {"User_Agent": "Mozilla/5.0"}
```

```python
req = Request(url, headers=headers)
raw_web = urlopen(req).read()
raw_web                                         # vemos que es variables de bits y no un string
```

```python
soup = BeautifulSoup(raw_web, "html.parser")    # Con requests, html.content donde html=requests.get(url)
```

### EXTRAEMOS TABLA DE WIKIPEDIA INSPECCIONANDO ELEMENTO

```python
tables = soup.find_all("table", attrs={"class": "wikitable sortable"})
# he quitado la ultima parte de "wikitable sortable jquery-tablesorter" y me aparecen 2 tablas (sino no aparecia ninguna)
table = tables[0]
```

```python
table_headers = []
for col in table.find_all("th")[1:]:        #quito el elemento [0] de "th" que es el titulo de la propia tabla
    table_headers.append(col.text.strip())
```

```python
table_body_raw = table.find_all("tr")[2:]
```

```python
table_body = []
i = 0
# hago lo de las i porque hay filas de año que se comparten con varias peliculas
for row in table_body_raw:

    template = []
    cols = row.find_all('td')
    if len(cols) == 3:
        template.append(table_body[i-1][0]) #fila anterior año
        for col in cols:
            template.append(col.text.strip()) #esta fila resto
    elif len(cols) == 2:
        template.append(table_body[i-1][0]) #fila anterior año
        template.append(cols[0].text.strip()) #fila anterior peli
        template.append(table_body[i-1][2]) #fila anterior personaje
        template.append(cols[1].text.strip()) #fila notas

    elif len(cols) == 4:
        for col in cols:
            template.append(col.text.strip()) #fila datos
    
    if len(template) == 4:
        table_body.append(template)
        i += 1
```
